```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
# Самостоятельно постройте парные регрессии от каждого атрибута с целевой переменной.
print("Парные регрессии:")
pairwise_results = []
for i in range(5):
    binary = LinearRegression().fit(x[[i]], y)
    y_pred = binary.predict(x[[i]])
    
    # Вычисляем метрики
    score = binary.score(x[[i]], y)
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    
    pairwise_results.append({
        "Признак": i,
        "R²": score,
        "Модель": "Линейная (парная)",
        "MSE": mse,
        "MAE": mae
    })
    
    print(f"Признак {i}: R² = {score:.4f}")
    
    plt.figure()
    plt.scatter(x[i], y)

    xx = np.linspace(x[i].min(), x[i].max(), 100).reshape((-1, 1)) # Строим 100 равномерных точек
    plt.plot(xx, binary.predict(xx), c='r')
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures

def polynomial_regression(x, y, degree):
    # Создание полиномиальных признаков
    poly = PolynomialFeatures(degree=degree)
    x_poly = poly.fit_transform(x)
    
    # Обучение линейной модели на полиномиальных признаках
    model = LinearRegression()
    model.fit(x_poly, y)
    
    # Предсказание и оценка
    y_pred = model.predict(x_poly)
    r2 = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    
    return model, poly, r2, mse, mae

# Для целевой переменной и каждого из признаков постройте полиномиальную регрессию второго, третьего и десятого порядков. Сделайте вывод о значимости регрессии.
poly_results = []
degrees = [2, 3, 10]
for i in x.columns:
    X_single = x[[i]]
    for d in degrees:
        model_poly, poly_transform, r2, mse, mae = polynomial_regression(X_single, y, degree=d)
        poly_results.append({
            "Признак": i,
            "Модель": f"Полиномиальная (d={d})",
            "R²": r2,
            "MSE": mse,
            "MAE": mae
        })
        print(f"Признак {i} (степень {d}): R² = {r2:.4f}, MSE = {mse:.4f}, MAE = {mae:.4f}")
        
        xx = np.linspace(X_single.min(), X_single.max(), 100).reshape(-1, 1)
        xx_poly = poly_transform.transform(xx)
        plt.figure(figsize=(6, 4))
        plt.scatter(X_single, y, color='blue', alpha=0.5, label="Данные")
        plt.plot(xx, model_poly.predict(xx_poly), color='red', label=f"Степень {d}")
        plt.title(f"Полиномиальная регрессия (степень {d}) для признака {i}")
        plt.xlabel("Признак 0")
        plt.ylabel("Целевая переменная")
        plt.legend()
        plt.show()
```

```python
# 4 Постройте сводную таблицу показателей точности всех построенных моделей. Сделайте вывод, какие модели переобученнные, какие - недообученные.
# Для линейных моделей (парных регрессий)
lin_results = pd.DataFrame(pairwise_results)
lin_results["Модель"] = "Линейная (парная)"

# Для полиномиальных моделей (собираем из списка poly_results)
poly_results_df = pd.DataFrame(poly_results)

# Объединяем таблицы
summary_df = pd.concat([lin_results, poly_results_df], sort=False).reset_index(drop=True)
print("\nСводная таблица показателей:")
print(summary_df)
```

```python
import time

# 6 (*) Замерьте время обучения полиномиальных моделей с разной степенью полинома. Сделайте вывод.

print("\nЗамер времени обучения полиномиальных моделей:")
time_results = []
test_degrees = [2, 3, 5, 10, 15]
X_sample = x
for d in test_degrees:
    start_time = time.time()
    poly = PolynomialFeatures(degree=d)
    X_poly = poly.fit_transform(X_sample)
    model = LinearRegression().fit(X_poly, y)
    elapsed = time.time() - start_time
    time_results.append((d, elapsed))
    print(f"Степень {d}: время обучения = {elapsed:.4f} секунд")

# Таблица с замерами времени
time_df = pd.DataFrame(time_results, columns=["Степень полинома", "Время обучения (сек)"])
print("\nВремя обучения для разных степеней полиномиальной регрессии:")
print(time_df)
```
